```python
df=df.dropna()
df=df.fillna("Not Available")
df
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df
```

```python
df['company_profile'] = df['company_profile'].str.lower()
```

```python
df['company_profile'] = df['company_profile'].str.strip()
df
```

```python
df['company_profile'] = df['company_profile'].str.replace('[^a-zA-Z ]','',regex=True)
```

```python
df[['min_salary','max_salary']] = df['salary_range'].str.split('-',expand=True)
```

```python
df['min_salary'] = pd.to_numeric(df['min_salary'], errors='coerce')
df['max_salary'] = pd.to_numeric(df['max_salary'], errors='coerce')
```

```python
df['job_type'] = df['fraudulent'].map({0:'Real',1:'Fake'})
df
```
